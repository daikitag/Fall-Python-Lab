# 1. Analyze COVID Data With Pie Charts

## 1-1. Import the required modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 1-2. Load the dataset

The number of cases are stored in "Race-COVID.csv" file in the "data" folder.
The
>pd.read_csv

function loads the csv file to Python as a dataframe.

```python
df = pd.read_csv('data/Race-COVID.csv')
```

## 1-3. Get information from the dataset

[ ] These square brackets are used to get a "column" of a dataframe. We specify the column name inside these square brackets.
>np.sum

funtion is used to add all the elements in the column to get the total number of cases.

```python
total = np.sum(df['Cases_Total'])
white = np.sum(df['Cases_White'])
black = np.sum(df['Cases_Black'])
latin = np.sum(df['Cases_LatinX'])
asian = np.sum(df['Cases_Asian'])
aian = np.sum(df['Cases_AIAN'])
nhpi = np.sum(df['Cases_NHPI'])
unknown = np.sum(df['Cases_Unknown'])
other = total - (white + black + latin + asian + aian + nhpi + unknown)

case = np.array([white, black, latin, asian, aian, nhpi, other, unknown])
```

## 1-4. Create pie chart

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = ['White','Black','Latin','Asian', 'AIAN', 'NHPI', 'Other', 'Unknown']

fig1, ax1 = plt.subplots(figsize=(5, 5))
ax1.pie(case, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)

ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('COVID Cases Broken by race/ethnicity',fontweight="bold", fontsize=15)
plt.show()

# NHPI stands for "Native Hawaiian and Pacific Islander"
# AIAN stands for "American Indian and Alaska Native"
```

Since there are many "unknown", it might be better to exclude it from the pie chart.

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = ['White','Black','Latin','Asian', 'AIAN', 'NHPI', 'Other']
case = np.array([white, black, latin, asian, aian, nhpi, other])

fig1, ax1 = plt.subplots(figsize=(5, 5))
ax1.pie(case, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)

ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('COVID Cases Broken by race/ethnicity',fontweight="bold", fontsize=15, pad=20)
plt.show()

# NHPI stands for "Native Hawaiian and Pacific Islander"
# AIAN stands for "American Indian and Alaska Native"
```

We will do the same thing for death cases

```python
total = np.sum(df['Deaths_Total'])
white = np.sum(df['Deaths_White'])
black = np.sum(df['Deaths_Black'])
latin = np.sum(df['Deaths_LatinX'])
asian = np.sum(df['Deaths_Asian'])
aian = np.sum(df['Deaths_AIAN'])
nhpi = np.sum(df['Deaths_NHPI'])
unknown = np.sum(df['Deaths_Unknown'])
other = total - (white + black + latin + asian + aian + nhpi + unknown)

case = np.array([white, black, latin, asian, aian, nhpi, other, unknown])
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = ['White','Black','Latin','Asian', 'AIAN', 'NHPI', 'Other', 'Unknown']

fig1, ax1 = plt.subplots(figsize=(5, 5))
ax1.pie(case, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)

ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Number of COVID death cases\n broken down by race/ethnicity',fontweight="bold", fontsize=15, pad=20)
plt.show()
```

## Summary of Pie Chart

## Advantages
- It is extremely easy to see the proportion of data
- You can just observe the sizes of circles to see the relative size of the data

## Disadvantages
- Pie charts can be difficult to see if too man pieces of data are used (AIAN and NHPI in this example)
- It can be difficult to compare multiple data, as we can't understand the size of the dataset

Before we go to the next step, I will add AIAN and NHPI to 'Other' and indicate how it can show the visualization of pie charts.

```python
total = np.sum(df['Deaths_Total'])
white = np.sum(df['Deaths_White'])
black = np.sum(df['Deaths_Black'])
latin = np.sum(df['Deaths_LatinX'])
asian = np.sum(df['Deaths_Asian'])
unknown = np.sum(df['Deaths_Unknown'])
other = total - (white + black + latin + asian + unknown)

case = np.array([white, black, latin, asian, other, unknown])
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = ['White','Black','Latin','Asian', 'Other', 'Unknown']

fig1, ax1 = plt.subplots(figsize=(5, 5))
ax1.pie(case, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)

ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Number of COVID death cases \n broken down by race/ethnicity',fontweight="bold", fontsize=15, pad=20)
plt.show()
```

It's much easier to see this pie chart, while loosing some information.

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
case = case
labels = ['White (48.2%)','Black (21.0%)','Latin (15.9%)','Asian (3.8%)',
          'AIAN (0.7%)', 'NHPI (0.1%)', 'Other (3.4%)', 'Unknown (6.9%)']

fig1, ax1 = plt.subplots(figsize=(5, 5))
pie = ax1.pie(case, shadow=True, startangle=90)

ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.legend(pie[0],labels, bbox_to_anchor=(1.3,0.5), loc="center right", fontsize=12, 
           bbox_transform=plt.gcf().transFigure)
plt.title('Number of COVID death cases \n broken down by race/ethnicity',fontweight="bold", fontsize=15)
plt.show()
```
