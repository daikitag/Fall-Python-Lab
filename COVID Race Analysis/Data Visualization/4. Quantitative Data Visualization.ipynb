```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Data/NHANES.csv')
adult = data[data['Age']>=18]

white = adult[adult['Race1'] == 'White']
black = adult[adult['Race1']=='Black']
hispanic = adult[adult['Race1']=='Hispanic']
asian = adult[adult['Race3']=='Asian']
other = adult[adult['Race1']=='Other']
```

```python
whiteBMI = white['BMI'].dropna()
blackBMI = black['BMI'].dropna()
hispanicBMI = hispanic['BMI'].dropna()
asianBMI = asian['BMI'].dropna()
otherBMI = other['BMI'].dropna()
```

```python
fig, axs = plt.subplots(figsize=(18,10))
axs.boxplot([whiteBMI,blackBMI,hispanicBMI,asianBMI,otherBMI])
plt.title('BMI of adult (+18) broken down by race', fontsize=20)
axs.set_xticklabels(['White', 'Black', 'Hispanic', 'Asian', 'Other'])
axs.set_ylabel('BMI', fontsize=18)
axs.tick_params(labelsize=15)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.hist(whiteBMI, bins=50)
ax.set_title('BMI of White Adult')
ax.set_xlabel('BMI')
ax.set_ylabel('freq')
fig.show()

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.hist(blackBMI, bins=50)
ax.set_title('BMI of Black Adult')
ax.set_xlabel('BMI')
ax.set_ylabel('freq')
fig.show()
```

```python
data['Income'] = data['HHIncomeMid']
data['Income'] = data['Income'].replace([2500, 7500],1)
data['Income'] = data['Income'].replace([12500, 17500, 22500],2)
data['Income'] = data['Income'].replace([30000, 40000, 50000],3)
data['Income'] = data['Income'].replace([60000, 70000, 87500],4)
data['Income'] = data['Income'].replace([100000],5)
```

```python
firstBMI = data[data['Income']==1]['BMI'].dropna()
secondBMI = data[data['Income']==2]['BMI'].dropna()
thirdBMI = data[data['Income']==3]['BMI'].dropna()
fourthBMI = data[data['Income']==4]['BMI'].dropna()
fifthBMI = data[data['Income']==5]['BMI'].dropna()
```

```python
fig, axs = plt.subplots(figsize=(18,10))
axs.boxplot([firstBMI, secondBMI, thirdBMI, fourthBMI, fifthBMI])
plt.title('BMI broken down by household income', fontsize=20)
axs.set_xticklabels(['First','Second','Third','Fourth','Fifth'])
axs.set_ylabel('BMI', fontsize=18)
axs.tick_params(labelsize=15)
plt.show()
```

First is the poorest and fifth is the richest
