### 2. 2017-2018 NHANES - Heights of Men in the U.S.
I took the latest version of the dataset from their website

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('Data/example.csv')
male_ID = data[data['RIAGENDR']==1].SEQN

df = pd.read_csv('Data/NHANES.csv')
male_df = df[df['SEQN'].isin(male_ID)] 
male_height = male_df['BMXHT']

kwargs = dict(hist_kws={'alpha':0.6}, kde_kws={'linewidth':2})

plt.figure(figsize=(10,7), dpi= 80)
sns.distplot(male_height, color="Blue",label=None, **kwargs)
plt.xlabel('Standing Height (cm)', size=15)
plt.ylabel('Proportion', size=15)
plt.title('Heights of Men in the U.S. 2017-2018 NHANES',weight='bold', size=15)
plt.legend();
```

### 2. The Shape of Natural Variation: The Normal Distribution

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('Data/blood.csv')
blood = data['BPXDI1']

kwargs = dict(hist_kws={'alpha':0.6}, kde_kws={'linewidth':2})

plt.figure(figsize=(10,7), dpi= 80)
sns.distplot(blood, color="Orange",label=None, **kwargs)
plt.xlabel('Diastolic: Blood pres (1st rdg) mm Hg', size=15)
plt.ylabel('Proportion', size=15)
plt.title('2017-2018 NHANES',weight='bold', size=15)
plt.legend();
```
