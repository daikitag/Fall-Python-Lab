### 1. Measles Cases and Outbreaks in 2019

```python
import matplotlib.pyplot as plt

year = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
cases = [63, 220, 55, 187, 667, 188, 86, 120, 375, 1282]

fig, ax = plt.subplots(figsize=(10,6))
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.bar(year, cases, align='center', alpha=0.5)
plt.grid(axis='y')
plt.xlabel('Year')
plt.ylabel('Number of Cases')
plt.ylim(0,1500)
plt.title('Measles Cases and Outbreaks in 2019')

for a,b in zip(year, cases):
    plt.text(a-0.2, b+10, str(b), color='b', size='large')

plt.show()
```

```python
import matplotlib.pyplot as plt

year = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
cases = [63, 220, 55, 187, 667, 188, 86, 120, 375, 1282]

fig, ax = plt.subplots(figsize=(10,6))
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.bar(year, cases, align='center', alpha=0.5)
plt.grid(axis='y')
plt.xlabel('Year')
plt.ylabel('Number of Cases')
plt.ylim(0,1500)
plt.title('Measles Cases and Outbreaks in 2019')
plt.plot(2019, 1282, 'r.', markersize=120, fillstyle='none')

for a,b in zip(year, cases):
    plt.text(a-0.2, b+10, str(b), color='b', size='large')

plt.show()
```
